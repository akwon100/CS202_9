```python
#Import
from bs4 import BeautifulSoup
import requests
#import urllib.parse
```

```python
#Letting server know we are a browser
def iAmBrowser(url):
    headers = requests.utils.default_headers()
    headers.update({'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',})
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content ,'lxml')
    return soup


url = 'https://www.imdb.com/list/ls064818015/'
soup = iAmBrowser(url)
```

```python
div_id = soup.find('div', id ="main")
```

```python
print(div_id)
```

```python
dclass = div_id.find_all('div', class_= "lister-item mode-detail")
```
